# **Pregunta 2**

¿Qué porcentaje de clientes tiene contratado servicio de internet y cómo influye este factor en
su gasto mensual promedio?, realice estadísticos y grafico adecuado. Interprete.

```python
# Creamos una variable booleana: True si tiene internet, False si no
df['TieneInternet'] = df['InternetService'] != 'No'

# 1) Porcentaje de clientes con servicio de internet
porcentaje_internet = df['TieneInternet'].mean() * 100

print(f"Porcentaje de clientes con servicio de internet: {porcentaje_internet:.2f}%\n")

# 2) Tabla de frecuencias (absolutas y relativas)
frecuencias = df['TieneInternet'].value_counts()
porcentajes = df['TieneInternet'].value_counts(normalize=True) * 100

tabla_frecuencias = pd.DataFrame({
    'TieneInternet': ['Con internet', 'Sin internet'],
    'Frecuencia': [frecuencias[True], frecuencias[False]],
    'Porcentaje': [porcentajes[True], porcentajes[False]]
})

print("Tabla de frecuencia de clientes con y sin internet:")
print(tabla_frecuencias, "\n")

# 3) Estadísticos de gasto mensual según tenga o no internet
estadisticos_gasto = df.groupby('TieneInternet')['MonthlyCharges'].agg(
    ['count', 'mean', 'median', 'std', 'min', 'max']
)

# Cambiamos el índice True/False por etiquetas más claras
estadisticos_gasto.index = ['Sin internet', 'Con internet'] if False in estadisticos_gasto.index else ['Con internet', 'Sin internet']

print("Estadísticos de MonthlyCharges según tenga o no internet:")
print(estadisticos_gasto, "\n")
```

```python
df['InternetCategory'] = np.where(df['TieneInternet'], 'Con internet', 'Sin internet')

# Gráfico 1: Porcentaje de clientes con y sin internet
plt.figure(figsize=(6,4))
ax = sns.countplot(x='InternetCategory', data=df)

plt.title("Distribución de clientes con y sin servicio de internet")
plt.xlabel("Tipo de cliente")
plt.ylabel("Cantidad de clientes")

# Añadimos los porcentajes sobre las barras
total = len(df)
for p in ax.patches:
    height = p.get_height()
    porcentaje = 100 * height / total
    ax.annotate(f'{porcentaje:.1f}%',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom')

plt.show()
```

```python
# Gráfico 2: Gasto mensual según tenga o no internet
plt.figure(figsize=(7,5))
sns.boxplot(x='InternetCategory', y='MonthlyCharges', data=df)
plt.title("Gasto mensual según servicio de internet")
plt.xlabel("Tipo de cliente")
plt.ylabel("Monthly Charges (USD)")
plt.show()
```

**Análisis e Interpretación**

El análisis muestra que el 78,98% de los clientes (3.727 personas) tiene contratado un servicio de internet, mientras que solo el 21,02% (992 clientes) no cuenta con este servicio. Esta diferencia marcada ya se observa en el primer gráfico de barras, donde la barra correspondiente a “Con internet” es aproximadamente cuatro veces más alta que la de “Sin internet”. Este predominio indica que el servicio de internet es un componente central dentro de los planes contratados y probablemente uno de los principales motores de ingresos para la empresa.
Desde el punto de vista económico, la diferencia entre ambos grupos es evidente. Los clientes sin internet presentan un gasto mensual muy bajo y homogéneo, con un promedio de 21,07 USD, una mediana de 20,15 USD y una desviación estándar de solo 2,17 USD. Además, los valores mínimos y máximos están muy acotados (18,70 USD – 26,90 USD). Todo esto indica que quienes no tienen internet suelen contratar planes básicos y similares entre ellos, con muy poca variabilidad en sus pagos mensuales.
En contraste, los clientes con internet muestran un comportamiento completamente distinto. Su gasto mensual promedio asciende a 76,63 USD, es decir, 3,6 veces más alto que el de los clientes sin internet. La mediana de 79,20 USD confirma que la mayoría de estos clientes paga montos elevados. Además, la desviación estándar de 21,83 USD es casi diez veces mayor que la del grupo sin internet, lo que indica una alta diversidad de planes, velocidades, servicios complementarios (como streaming, seguridad o respaldo online) y posibles promociones.

En cuanto a la fidelidad alta, el Percentil 75 alcanza los 55 meses, evidenciando que el 25% superior de la base de clientes mantiene una relación comercial de largo plazo (más de 4 años y medio).

# **Pregunta 5**
¿Cómo se relacionan los cargos mensuales con el total cobrado, ¿Existe una correlación
significativa?, use gráficas y tablas estadísticas. Interprete.

```python
print("\n--- 5.1 Cálculo de Correlación (Pearson) ---")
matriz_correlacion = df[['MonthlyCharges', 'TotalCharges']].corr()

print("Matriz de Correlación:")
print(matriz_correlacion)

r_valor = matriz_correlacion.loc['MonthlyCharges', 'TotalCharges']
print(f"\nCoeficiente de Correlación (r): {r_valor:.4f}")
```

```python

print("\n--- 5.2 Tabla de Relación: Gasto Mensual vs. Total Promedio ---")

df['Rango_Mensual'] = pd.qcut(df['MonthlyCharges'], q=4, labels=['Bajo', 'Medio', 'Alto', 'Muy Alto'])


tabla_contexto = df.groupby('Rango_Mensual', observed=False)['TotalCharges'].mean().reset_index()
tabla_contexto.columns = ['Nivel de Gasto Mensual', 'Total Cobrado Promedio ($)']
print(tabla_contexto)
```

```python

print("\n--- 5.3 Gráfico de Dispersión (Scatterplot) ---")
plt.figure(figsize=(10, 6))
sns.regplot(x='MonthlyCharges', y='TotalCharges', data=df,
            line_kws={"color": "red"},
            scatter_kws={"alpha": 0.5})
plt.title('Relación entre Cargos Mensuales y Totales')
plt.xlabel('Cargos Mensuales ($)')
plt.ylabel('Cargos Totales ($)')
plt.grid(True, alpha=0.3)
plt.show()
```

**Análisis e Interpretación**

Al estudiar la relación entre los cargos mensuales y el total facturado histórico, el coeficiente de correlación de Pearson presentado en la sección 5.1 arroja un valor de r = 0.65 (aproximado). Esto indica estadísticamente una correlación positiva significativa: cuando una variable aumenta, la otra también tiende a hacerlo.

Esta relación se confirma numéricamente en la Tabla de Relación (5.2). Allí se evidencia que los clientes en el rango de gasto mensual "Bajo" tienen un total acumulado promedio mucho menor que los clientes del rango "Muy Alto". Sin embargo, el Gráfico de Dispersión (5.3) muestra que la relación no es una línea perfecta (los puntos están dispersos en forma de abanico). Esto ocurre porque el TotalCharges depende también del tiempo de permanencia (tenure), un cliente puede pagar mucho mensualmente, pero si es nuevo, su total acumulado será bajo, lo que explica la dispersión observada en el gráfico.

# Pregunta 6

Entre los clientes con contrato de un año y los de dos años, ¿quién presenta mayor
variabilidad en el gasto mensual?, use la medida estadística adecuada y gráfico adecuado.


```python
contratos_largos_df = df[df['Contract'].isin(['One year', 'Two year'])].copy()

print("--- 6.1 Estadísticos Descriptivos (Contratos de 1 y 2 años) ---")

estadisticos_contratos_largos = contratos_largos_df.groupby('Contract')['MonthlyCharges'].describe()
print(estadisticos_contratos_largos)

print("\n--- 6.2 Coeficiente de Variación (CV) ---")

std_dev = estadisticos_contratos_largos['std']
mean_val = estadisticos_contratos_largos['mean']

cv = (std_dev / mean_val) * 100

print("El Coeficiente de Variación (en %) para el gasto mensual es:")
print(cv)

```

**Análisis e Interpretación**

Para determinar qué grupo de clientes de los contratos de 'One Year' y 'Two Year' presentan una mayor variabilidad en el gasto mensual, se utilizó como medida estadística principal el Coeficiente de Variación (CV). Esta medida es la más adecuada ya que permite comparar la dispersión de dos grupos, incluso si sus promedios fuesen diferentes (en este caso son: 'One Year' -> 65.848203 y 'Two Year' -> 59.857239).

Al analizar los datos de la consola, se obtienen los siguientes resultados:

*   Coeficiente de Variación (CV) 'One year': 47.76%
*  Coeficiente de Variación (CV) 'Two year': 58.08%

Este resultado también es coherente con la desviación estándar (std), donde el grupo 'One year' tiene una std de 31.45 y el grupo 'Two year' tiene una std de 34.76.

```python
print("\n--- 6.3 Generando Gráfico Boxplot Comparativo ---")
plt.figure(figsize=(10, 7))
sns.boxplot(x='Contract', y='MonthlyCharges', data=contratos_largos_df)
plt.title('Distribución del Gasto Mensual (Contratos de 1 y 2 años)')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Gasto Mensual (USD)')
plt.show()
```



**En relacion al gráfico**

El gráfico de caja "Distribución del Gasto Mensual (Contratos de 1 y 2 años)" ilustra esta diferencia de forma clara. Se observa visualmente que la caja azul ('Two year') es significativamente más "alta" que la caja naranja ('One year'). Esto indica que el 50% central de los clientes de dos años tiene gastos mucho más dispersos. Además, los "bigotes" (las líneas que salen de la caja) del grupo 'Two year' cubren un rango de valores más amplio, confirmando que sus gastos están más dispersos.

**Interpretación**

Los resultados indican de forma concluyente que los clientes con contrato de dos años ('Two year') presentan una mayor variabilidad en su gasto mensual. A pesar de que su gasto promedio es menor (59.85 USD) que el de los clientes de un año (65.84 USD), sus gastos son menos consistentes y más impredecibles. Esto sugiere que dentro del grupo de 'Dos años' existe una mayor diversidad de perfiles de consumo: probablemente incluye un gran grupo de clientes que contrata servicios muy básicos (quizás con descuentos por lealtad, lo que baja el promedio) y, al mismo tiempo, un grupo de clientes que contrata los paquetes más completos y caros (lo que explica la alta desviación estándar y el alto CV).

En resumen, el gasto de los clientes de 'One 'One year' es más predecible y está más concentrado alrededor de su promedio

# **Pregunta 8**
Calcula el coeficiente de variación del gasto mensual para clientes con servicio de internet y
para aquellos sin servicio. ¿Qué nos indica esta comparación?

```python
# Coeficiente de Variación (CV)

# Filtrar grupos
grupo_con_internet = df[df['TieneInternet'] == True]['MonthlyCharges']
grupo_sin_internet = df[df['TieneInternet'] == False]['MonthlyCharges']

# Calcular CV = desviación estándar / media
cv_con_internet = grupo_con_internet.std() / grupo_con_internet.mean()
cv_sin_internet = grupo_sin_internet.std() / grupo_sin_internet.mean()

print(f"Coeficiente de variación (Con internet): {cv_con_internet:.4f}")
print(f"Coeficiente de variación (Sin internet): {cv_sin_internet:.4f}")
```

**Análisis e Interpretación**

Los clientes sin internet presentan un coeficiente de variación de 0.1028, lo que indica que su gasto mensual es muy homogéneo y varía muy poco entre personas. Esto se debe a que todos pagan prácticamente un mismo tipo de plan básico, con precios similares y poca dispersión.
En cambio, los clientes con internet tienen un coeficiente de variación de 0.2849, casi tres veces mayor. Esto significa que su gasto mensual es mucho más diverso, ya que existen múltiples planes, velocidades y servicios adicionales que generan diferencias importantes en los montos pagados por cada usuario.
En resumen, el gasto mensual de los clientes sin internet es estable y uniforme, mientras que el de los clientes con internet es más variado y depende fuertemente del tipo de plan contratado.

**Interpretación del gráfico**

El gráfico muestra patrones muy claros y distintos para cada tipo de contrato:

# **Pregunta 11**
¿Qué método de pago es el más común y cómo se relaciona con el gasto mensual?



```python
# 1) Método de pago más común
metodo_mas_comun = df['PaymentMethod'].value_counts().idxmax()
print("Método de pago más común:", metodo_mas_comun)

print("\nFrecuencias de métodos de pago:")
print(df['PaymentMethod'].value_counts())

# 2) Gasto mensual promedio por método de pago
promedio_por_metodo = df.groupby('PaymentMethod')['MonthlyCharges'].mean().sort_values(ascending=False)
print("\nGasto mensual promedio por método de pago:")
print(promedio_por_metodo)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='PaymentMethod', y='MonthlyCharges', data=df)
plt.title("Gasto Mensual según Método de Pago")
plt.xlabel("Método de Pago")
plt.ylabel("Monthly Charges (USD)")
plt.xticks(rotation=45)
plt.show()
```

**Análisis e Interpretación**

El método de pago más común entre los clientes es Electronic check, con 1.567 usuarios, lo que lo convierte en la forma de pago predominante en la empresa. Al comparar el gasto mensual entre los distintos métodos, se observa que quienes utilizan Electronic check también presentan el gasto promedio más alto, con 76,27 USD, lo que sugiere que este grupo tiende a contratar planes más costosos o agregar servicios adicionales. En cambio, los clientes que pagan mediante Mailed check muestran el gasto promedio más bajo (44,69 USD), reflejando planes más económicos. Los métodos automáticos (Bank transfer y Credit card) se sitúan en un nivel intermedio.
En conclusión, el método de pago más común es Electronic check, y este grupo destaca por tener los gastos mensuales más elevados, indicando que existe una relación entre este método de pago y la contratación de planes de mayor valor.
